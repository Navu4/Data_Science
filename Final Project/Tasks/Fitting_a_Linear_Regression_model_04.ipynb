# Import the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
```

# Import Dataset

```python
#Read files:
df = pd.read_csv("clean_2_train.csv")
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.info()
```

#### Created a list remove_cols to remove those columns which doesn't required for Model Building 

```python
remove_cols = [
    'Item_Identifier',
    'Item_Type',
    'Outlet_Identifier',
    'Outlet_Establishment_Year'
]
df = df.drop(remove_cols,axis =1)
df.head()
```

```python
df.shape
```

# Correlation Matrix

```python
plt.figure(figsize = (20,20))
sns.heatmap(df.corr(),annot = True)
```

## Data Preprocessing

```python
y = df.Item_Outlet_Sales.values
X = df.drop('Item_Outlet_Sales',axis = 1)
```

```python
print(X.shape,y.shape)
```

```python
print(X[:5])
```

```python
print(y[:5])
```

###  Using scikit-learn's train_test_split to make a new set of training and testing data.  

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y ,test_size = 0.2,random_state = 2 )
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

### Trained a Linear Regression model using the training data from your split.

```python
from sklearn.linear_model import LinearRegression
#Training the Simple Linear Regression model on the Training set
model = LinearRegression(normalize=True)
```

______________________________________________________________________
**Now, train on the training data and obtain predicted classes, as well as class probabilities, using the testing data.**

```python
model.fit(X_train, y_train)
```

```python
#Predicting the Test set results
y_pred = model.predict(X_train)
```

```python
from sklearn.metrics import r2_score,mean_squared_error

score = r2_score(y_train,y_pred)
print("Score of Training:",100*score)
print("RMSE : %.4g" % np.sqrt(mean_squared_error(y_train,y_pred)))
```

```python
from sklearn.metrics import mean_squared_error,make_scorer
from sklearn.model_selection import cross_val_score

#Perform cross-validation:
cv_score = cross_val_score(model,X_train, y_train, cv=20, scoring = make_scorer(mean_squared_error))
cv_score = np.sqrt(np.abs(cv_score))
    
#Print model report:
print("\nModel Report")
print("RMSE : %.4g" % np.sqrt(mean_squared_error(y_train,y_pred)))
print("CV Score : Mean - %.4g | Std - %.4g | Min - %.4g | Max - %.4g" % (np.mean(cv_score),np.std(cv_score),np.min(cv_score),np.max(cv_score)))
```

```python
y_test_pred = model.predict(X_test)
```

```python
score = r2_score(y_test,y_test_pred)
print("Score of Testing:",100*score)
print("RMSE : %.4g" % np.sqrt(mean_squared_error(y_test,y_test_pred)))
```

```python
#Perform cross-validation:
cv_score = cross_val_score(model,X_test, y_test, cv=20, scoring = make_scorer(mean_squared_error))
cv_score = np.sqrt(np.abs(cv_score))
    
#Print model report:
print("\nModel Report")
print("RMSE : %.4g" % np.sqrt(mean_squared_error(y_test,y_test_pred)))
print("CV Score : Mean - %.4g | Std - %.4g | Min - %.4g | Max - %.4g" % (np.mean(cv_score),np.std(cv_score),np.min(cv_score),np.max(cv_score)))
```

```python
model.coef_
```

```python
coef2 = pd.Series(model.coef_ ,X_train.columns).sort_values()
plt.figure(figsize = (10,10))
coef2.plot(kind='bar', title='Model Coefficients')
```

# Continuing Verification of Data Integrity

With our knowledge of Boolean arrays, we will examine some of the duplicate IDs

```python
#Let's check ID column:
df['ID']
```

```python
#Calculating the occurence of IDs in the dataset 
#with help of values_counts()
#We have 29687 unqiue values
id_counts = df['ID'].value_counts()
id_counts
```

```python
#Dupe_mask stores true value of IDs whether the ID has occur twice or not 
#Setting ID with boolean masking in our dataframe
dupe_mask = id_counts == 2
dupe_mask
```

```python
#dupe_ids stores only duplicate IDs of the dataset i.e the IDs with the True value in dupe_mask which are only 313
dupe_ids = id_counts.index[dupe_mask]
print(dupe_ids)
```

```python
#Storing the duplicates IDs in dupe_ids
#Converting the dupe_ids dataframe into a list
#Calculating the total number of duplicate entries
dupe_ids = id_counts.index[dupe_mask]
dupe_ids = list(dupe_ids)
len(dupe_ids)
```

```python
#Examining the data for duplicate IDs
#Series dataframe in which we are storing the truth value of the IDs present or not in our orignial dataset 
df['ID'].isin(dupe_ids[0:3])
```

```python
#Displaying only first 3(three) accounts which has occured twice in the dataset
#In general we have noticed that there are 313 duplicate entries
df.loc[df['ID'].isin(dupe_ids[0:3]),:].head(10)
```

### Implementing OHE for a Categorical Feature

```python
# Create an empty column for the categorical labels called EDUCATION_CAT
df_clean_2['EDUCATION_CAT'] = 'none'
```

```python
# Examine the first few rows of the DataFrame for the EDUCATION and EDUCATION_CAT columns
df_clean_2[['EDUCATION', 'EDUCATION_CAT']].head(10)
```

```python
# Create a dictionary that describes the mapping for education categories
# Mapping the value of EDUCATION_CAT according to the EDUCATION feature
# where  1: "graduate school",2: "university",3: "high school" and 4: "others"
cat_mapping = {
    1: "graduate school",
    2: "university",
    3: "high school",
    4: "others"
}
```

```python
# Apply the mapping to the original EDUCATION column using .map and assign the result to the new EDUCATION_CAT column
df_clean_2['EDUCATION_CAT'] = df_clean_2['EDUCATION'].map(cat_mapping)
df_clean_2[['EDUCATION', 'EDUCATION_CAT']].head(10)
```

Now we are ready to one-hot encode. We can do this by passing a Series of a DataFrame to the pandas get_dummies() function. The function got this name because one-hot encoded columns are also referred to as dummy variables. The result will be a new DataFrame, with as many columns as there are levels of the categorical variable.

```python
# Create a one-hot encoded DataFrame of the EDUCATION_CAT column. Examine the first 10 rows:
edu_ohe = pd.get_dummies(df_clean_2['EDUCATION_CAT'])
edu_ohe.head(10)
```

```python
# Concatenate the one-hot encoded DataFrame to the original DataFrame
df_with_ohe = pd.concat([df_clean_2, edu_ohe], axis=1)
df_with_ohe[['EDUCATION_CAT', 'graduate school',
             'high school', 'university', 'others']].head(10)
```

```python
# Write the latest DataFrame to a CSV (comma-separated value) file
df_with_ohe.to_csv('cleaned_data.csv', index=False)
```
